Задание №2
===

```python
dau_channel.head()
```

```python
dau_channel['BrainPlus'].median() , dau_channel['GameAds'].median() , dau_channel['JetAds'].median() , dau_channel['SenseAds'].median()
```

```python
dau_channel['Total'] = dau_channel['BrainPlus'] + dau_channel['GameAds'] + dau_channel['JetAds'] + dau_channel['SenseAds']
dau_channel.head()
```

```python
dau_channel_percent = dau_channel
dau_channel_percent['BrainPlus'] =  (dau_channel['BrainPlus'] / dau_channel['Total'] * 100)
dau_channel_percent['GameAds'] =  (dau_channel['GameAds'] / dau_channel['Total'] * 100)
dau_channel_percent['JetAds'] =  (dau_channel['JetAds'] / dau_channel['Total'] * 100)
dau_channel_percent['SenseAds'] =  (dau_channel['SenseAds'] / dau_channel['Total'] * 100)
dau_channel_percent['Total'] =  100
dau_channel_percent.head()
```

```python
dau_channel_percent[['BrainPlus', 'GameAds', 'JetAds', 'SenseAds']].plot.area(figsize=(15, 5))
plt.title('Изменение DAU')
plt.xlabel('Дата')
plt.ylabel('Уникальные пользователи')
```

Исходя из полученных данных можно сделать вывод, что BrainPlus приносит около 50 % всех пользователей. GameAds и JetAds приносят сравнительно одинаковое количество пользователей. SenseAds на протяжении всего периода приносит меньше пользователей, чем остальные

```python
mau_region = pd.read_excel('hw_lesson03_mau_region.xlsx')
mau_region.info(5)
```

```python
mau_region.set_index('month', inplace=True)
mau_region.head()
```

```python
mau_region.loc[5, 'Africa'] = 5000
mau_region.head()
```

```python
def calculate_mau_region(row):
    total_mau = row['Australia'] + row['Africa']
    if total_mau > row['CIS']:
        return total_mau
    else:
        return 0
mau_region.apply(calculate_mau_region, axis=1)
```

```python
mau_region['EU превышает порог'] = mau_region['Europe'] >= 4000
mau_region.head()
```

Задание №4
===

```python
def super_func(file_name):
    dau_func = pd.read_excel(file_name)
    dau_func.set_index('date', inplace=True)
    print(dau_func)
    dau_func.plot()
    plt.xlabel('Дата')
    plt.ylabel('Уникальные посетители')
    plt.ylim(0)
    dau_percentage = dau_func.copy()
    total_dau_per_date = dau_percentage.sum(axis=1)
    dau_percentage = dau_percentage.div(total_dau_per_date, axis=0) * 100
    dau_percentage.plot.area(figsize=(15, 5))
    plt.xlabel('Дата')
    plt.ylabel('% уникальных посетителей')

print(super_func('hw_lesson03_dau_region.xlsx'))
```

```python
print(super_func('hw_lesson03_dau_channel.xlsx'))
```

```python
print(super_func('hw_lesson03_dau_device.xlsx'))
```
